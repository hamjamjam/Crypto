```python
from time import sleep
#create base url for Alpha Vantage API
url = 'https://www.alphavantage.co/query?'

#loop through all symbols
data = {} 
for symbol in symbols:
    print("Getting Data for {}".format(symbol))
    params = {'function': 'DIGITAL_CURRENCY_DAILY',
              'symbol': symbol,
              'market': 'USD',
              'apikey': 'JIFTM7UPB8DLKW9I'}
    
    response = ''
    #try to find historical data for the given cryptocurrency
    try:
        response = requests.get(url=url, params=params)
        print(response)
    except:
        #Alpha Vantage does not have data for this asset
        sleep(60)
        response = requests.get(url=url, params=params)
        print(response)
    
    output = response.json()
        
    #turn output into a pandas dataframe and format it correctly
    ts_data = output[list(output.keys())[1]]
    df = pd.DataFrame.from_dict(ts_data)
    df = df.T
    df = df.drop(columns=['1b. open (USD)', '2b. high (USD)', '3b. low (USD)', '4b. close (USD)'])
    df = df.rename(columns={'1a. open (USD)': 'open', '2a. high (USD)': 'high', '3a. low (USD)': 'low', '4a. close (USD)': 'close', '6. market cap (USD)': 'market cap'})

    #append new asset time series data to dictionary
    data[symbol] = df  
    print("Data Length: {}\n".format(len(data)))
        
    sleep(2.5)
```

```python
for key in data:
    filename = "Folder/{}.csv".format(key)
    df = data[key]
    df.to_csv(filename, index=False)   
```
